```python
class Person:
    def __init__(self, name, dept):
        self.name = name
        self.dept = dept
    
    def get_details(self):
        return f"Name: {self.name} Department:{self.dept}"
```

```python
p1 = Person("Pragat", "CS")
```

```python
p1.get_details()
```
